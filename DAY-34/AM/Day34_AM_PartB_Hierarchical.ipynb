# Day 34 AM – Part B: Hierarchical Clustering

Agglomerative clustering and dendrogram on Iris.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering, KMeans
from sklearn.metrics import adjusted_rand_score
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
iris = load_iris()
X = iris.data
y = iris.target

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
agg = AgglomerativeClustering(n_clusters=3)
agg_labels = agg.fit_predict(X_scaled)
agg_ari = adjusted_rand_score(y, agg_labels)

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans_labels = kmeans.fit_predict(X_scaled)
kmeans_ari = adjusted_rand_score(y, kmeans_labels)

print("Agglomerative ARI:", round(agg_ari, 4))
print("K-Means ARI:", round(kmeans_ari, 4))

if agg_ari > kmeans_ari:
    print("Agglomerative clustering gives better ARI.")
elif agg_ari < kmeans_ari:
    print("K-Means gives better ARI.")
else:
    print("Both methods give the same ARI.")
```

```python
Z = linkage(X_scaled, method="ward")
plt.figure(figsize=(12, 5))
dendrogram(Z)
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Sample Index")
plt.ylabel("Distance")
plt.tight_layout()
plt.savefig("iris_dendrogram.png", dpi=150)
plt.show()
```
