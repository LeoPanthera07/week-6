Personal ML quick reference for Week 6 with code snippets, use cases, hyperparameters, flowchart, and small verification on Wine dataset.

````python
import numpy as np
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
````

## Logistic Regression

**Description:** Linear classifier for probability-based decision boundaries.

**When to use:** Use when features and target relation is mostly linear.

**Key hyperparameters:** C, solver

**3-line code example:**
```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(C=1.0, max_iter=1000)
model.fit(X, y)
```

## Decision Tree

**Description:** Rule-based supervised model that splits features greedily.

**When to use:** Use when interpretability matters.

**Key hyperparameters:** max_depth, min_samples_split

**3-line code example:**
```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_depth=5, random_state=42)
model.fit(X, y)
```

## Random Forest

**Description:** Ensemble of many trees trained with bagging.

**When to use:** Use for strong tabular-data baseline.

**Key hyperparameters:** n_estimators, max_depth

**3-line code example:**
```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X, y)
```

## AdaBoost

**Description:** Boosting model that focuses on earlier mistakes.

**When to use:** Use for simple boosted baseline on structured data.

**Key hyperparameters:** n_estimators, learning_rate

**3-line code example:**
```python
from sklearn.ensemble import AdaBoostClassifier
model = AdaBoostClassifier(n_estimators=100, learning_rate=0.5, random_state=42)
model.fit(X, y)
```

## XGBoost

**Description:** Gradient boosting model with strong regularization and speed.

**When to use:** Use when you need very high accuracy on tabular data.

**Key hyperparameters:** n_estimators, max_depth

**3-line code example:**
```python
import xgboost as xgb
model = xgb.XGBClassifier(n_estimators=200, max_depth=4, random_state=42)
model.fit(X, y)
```

## LightGBM

**Description:** Histogram-based gradient boosting model optimized for speed.

**When to use:** Use for large tabular datasets.

**Key hyperparameters:** n_estimators, num_leaves

**3-line code example:**
```python
import lightgbm as lgb
model = lgb.LGBMClassifier(n_estimators=200, num_leaves=31, random_state=42)
model.fit(X, y)
```

## Voting

**Description:** Combines predictions from multiple base models.

**When to use:** Use when several decent models complement each other.

**Key hyperparameters:** estimators, voting

**3-line code example:**
```python
from sklearn.ensemble import VotingClassifier
model = VotingClassifier(estimators=[('lr', lr), ('rf', rf)], voting='soft')
model.fit(X, y)
```

## Stacking

**Description:** Learns a meta-model on top of base-model predictions.

**When to use:** Use when you want stronger ensembles from diverse models.

**Key hyperparameters:** estimators, final_estimator

**3-line code example:**
```python
from sklearn.ensemble import StackingClassifier
model = StackingClassifier(estimators=[('lr', lr), ('rf', rf)], final_estimator=LogisticRegression())
model.fit(X, y)
```

## SVM

**Description:** Margin-based classifier that can use kernels for non-linear boundaries.

**When to use:** Use for small or medium datasets with clean separation.

**Key hyperparameters:** C, gamma

**3-line code example:**
```python
from sklearn.svm import SVC
model = SVC(kernel='rbf', C=1.0, gamma='scale')
model.fit(X, y)
```

## KNN

**Description:** Instance-based method that predicts using nearest neighbors.

**When to use:** Use for small datasets with local structure.

**Key hyperparameters:** n_neighbors, weights

**3-line code example:**
```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=5, weights='uniform')
model.fit(X, y)
```

## K-Means

**Description:** Unsupervised clustering using centroid updates.

**When to use:** Use for compact clusters when k is known or estimated.

**Key hyperparameters:** n_clusters, n_init

**3-line code example:**
```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3, random_state=42, n_init=10)
labels = model.fit_predict(X)
```

## DBSCAN

**Description:** Density-based clustering with noise detection.

**When to use:** Use for irregular-shaped clusters and outlier detection.

**Key hyperparameters:** eps, min_samples

**3-line code example:**
```python
from sklearn.cluster import DBSCAN
model = DBSCAN(eps=0.5, min_samples=5)
labels = model.fit_predict(X)
```

## PCA

**Description:** Dimensionality reduction by maximizing explained variance.

**When to use:** Use before visualization or when features are highly correlated.

**Key hyperparameters:** n_components, svd_solver

**3-line code example:**
```python
from sklearn.decomposition import PCA
model = PCA(n_components=2)
X_reduced = model.fit_transform(X)
```

## Algorithm Selection Flowchart

```text
Start
├── Need supervised prediction?
│   ├── Need interpretability?
│   │   ├── Yes -> Logistic Regression or Decision Tree
│   │   └── No -> Random Forest / AdaBoost / XGBoost / LightGBM
│   ├── Small or medium data with clear margin? -> SVM
│   ├── Small data with local patterns? -> KNN
│   └── Want ensemble of many models? -> Voting or Stacking
└── Need unsupervised structure?
    ├── Want fixed number of clusters? -> K-Means
    ├── Want noise detection / irregular shapes? -> DBSCAN
    └── Want lower-dimensional representation? -> PCA
```

## Verification on Wine Dataset

````python
wine = load_wine()
X = wine.data
y = wine.target

models = {
    'Logistic Regression': Pipeline([('scaler', StandardScaler()), ('model', LogisticRegression(max_iter=1000, random_state=42))]),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'SVM': Pipeline([('scaler', StandardScaler()), ('model', SVC(kernel='rbf', C=1.0, gamma='scale'))])
}

rows = []
for name, model in models.items():
    scores = cross_val_score(model, X, y, cv=5, scoring='accuracy')
    rows.append({'Model': name, 'Mean Accuracy': scores.mean(), 'Std': scores.std()})

results = pd.DataFrame(rows).sort_values('Mean Accuracy', ascending=False)
print(results.round(4).to_string(index=False))
````
